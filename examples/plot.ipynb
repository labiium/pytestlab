```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load("data/anthropic.claude-3-5-sonnet-20241022-v2:0_exp1.npy")

plt.figure(figsize=(10, 6))  # Adjust figure size as needed

print(data.shape)
print(data[0, 0, :])
for i in range(3):
    plt.plot(data[i, :, 1], data[i, :, 2], marker='o',label=f"Data {data[i, 0, 0]}")

plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Data Sweep")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load("data/adaptive_sweep_FAST_10.npy")

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assume `data` is the 2D array with shape (100, 3) containing [X, Y, Z] coordinates
# Example:
# data = np.array([[X1, Y1, Z1], [X2, Y2, Z2], ...])


# Step 1: Separate X, Y, and Z from `data`
X = data[:, 0]
Y = data[:, 1]
Z = data[:, 2]

# Step 2: Reshape X, Y, and Z into grids
# Assuming data is a 10x10 grid (adjust shape based on actual data structure)
grid_size = int(np.sqrt(data.shape[0]))  # Calculate grid size, e.g., 10 if 100 points

X = X.reshape(grid_size, grid_size)
Y = Y.reshape(grid_size, grid_size)
Z = Z.reshape(grid_size, grid_size)

# Step 3: Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Step 4: Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')  # Customize colormap if needed

# Optional: Add labels
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

plt.show()
```
